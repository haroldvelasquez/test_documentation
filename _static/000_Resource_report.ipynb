```python
# import sys
# sys.path.insert(0,r'C:\HVS\Devs\Python\packages\automod\package')
import automod as ct
import pandas as pd
import numpy as np
```

```python
mined_out = {  
    'MinedCode': [[0]]
}

classif_code = {
    'ClassCode':[[1],[2],[1,2],[3],[1,2,3]],
}

otype_code = {
    'TypeCode':[['Oxd'],['Sul'], ['Oxd', 'Sul']],
}

vars = {
    'SEN02_Ag':{'unit':'gpt'},
    'SEN02_Au':{'unit':'gpt'},
    'SEN02_Pb':{'unit':'pct'},
    'SEN02_Zn':{'unit':'pct'},
    'SEN02_AgEq':{'unit':'gpt'},
} 

cutoff_col = 'SEN02_AgEq'
density_col = 'SG'
cutoff = 135
```

```python
df = pd.read_csv(f'SRK_SanNicolas_OCTR.csv')

model = ct.BM(df=df,x='X',y='Y',z='Z', dx='dX',dy='dY',dz='dZ')
report = model.report(
            vars=vars,
            cutoff=cutoff, 
            cutoff_col=cutoff_col,
            density_col=density_col,
            mined_out=mined_out,
            classif_code=classif_code,
            otype_code=otype_code
            )

#report.to_csv('resource_report.csv',index=False)
report
```
